# Step 08: Define Preprocessing Technique

preprocess.png

**Stop words**: the, of, and, to, ...  

* Typically about **400 to 500 such words**
* Accounts **20-30% of total word counts**.

**Stemming**: to improve effectiveness by combining many words with the same roots into 1 word i.e. use for (user, using, usable, users, used, etc)

* matching **similar words**
* mainly improve **recall**
* may reduce indexing size as much as **40-50%**  



```python
stop_words = list(set(stopwords.words('english')))
valid_words = set(word.lower() for word in words.words())
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

def preprocess(text, remove_stop_words=False, stem=False, lemmatize=False):
    """
        Input:
            text: String       --> feature text.
            stemming: Boolean  --> T/F whether to use tf_idf technique or not
            lemmatize: Boolean --> T/F whether to use tf_idf technique or not
        Output:
            text: String       --> preprocessed text
    """
    text = text.lower()                           # normalize
    tokens = word_tokenize(text) #text.split()    # tokenize
    words = [word for word in tokens if word in valid_words] # take only valid words

    if remove_stop_words:
        words = [word for word in words if word not in stop_words]

    if stem:
        words = [stemmer.stem(word) for word in words]

    if lemmatize:
        words = [lemmatizer.lemmatize(word) for word in words]

    return " ".join(words)
```

# Step 09: Define Different Combinations of Preprocessing Techniques

```python
preprocessing_vocab = {
    "text_000": {"remove_stop_words": False, "stem": False, "lemmatize": False},
    "text_001": {"remove_stop_words": False, "stem": False, "lemmatize": True},
    "text_010": {"remove_stop_words": False, "stem": True, "lemmatize": False},
    "text_011": {"remove_stop_words": False, "stem": True, "lemmatize": True},
    "text_100": {"remove_stop_words": True, "stem": False, "lemmatize": False},
    "text_101": {"remove_stop_words": True, "stem": False, "lemmatize": True},
    "text_110": {"remove_stop_words": True, "stem": True, "lemmatize": False},
    "text_111": {"remove_stop_words": True, "stem": True, "lemmatize": True},

}


def getPreprocessingConfig(name):
    configured_name = ""
    for key, value in preprocessing_vocab[name].items():
        if configured_name:
            configured_name += "-"
        configured_name += f"{key}={value}"

    return configured_name
```

# Step 10: Develop Dataset Using All Above Defined Preprocessing Techniques

```python
for technique, args in preprocessing_vocab.items():
    df[technique] = df['text'].apply(lambda x: preprocess(x, **args))
```

# Step 11: Implement MLModel Class for (Training, Testing & Evaluation)

```python
class MLModel:
    def __init__(self, model, train_x, train_y, test_x, test_y, stemmer, lemmatizer, tokenizer, vectorizer):
        self.model = model
        self.train_x = train_x
        self.train_y = train_y
        self.test_x = test_x
        self.test_y = test_y
        self.stemmer = stemmer
        self.lemmatizer = lemmatizer
        self.tokenizer = tokenizer
        self.vectorizer = vectorizer
        self.stop_words = list(set(stopwords.words('english')))
        self.valid_words = set(word.lower() for word in words.words())
        self.results = {}


    def preprocess(self, text, remove_stop_words=False, stem=False, lemmatize=False):
        text = text.lower()                # normalize
        tokens = self.tokenizer(text)      # tokenize
        tokens = [
            token for token in tokens if token in self.valid_words and token.isalpha()
        ] # take only valid words

        # remove stop words
        if remove_stop_words:
            tokens = [token for token in tokens if token not in self.stop_words]

        if stem:
            tokens = [self.stemmer.stem(token) for token in tokens]

        if lemmatize:
            tokens = [self.lemmatizer.lemmatize(token) for token in tokens]

        return " ".join(tokens)

    def train(self):
        train_x_vec = self.vectorizer.fit_transform(self.train_x)
        self.model.fit(train_x_vec, self.train_y)


    def test(self):
        test_x_vec = self.vectorizer.transform(self.test_x)
        y_hat = self.model.predict(test_x_vec)
        accuracy = accuracy_score(self.test_y, y_hat)
        precision = precision_score(self.test_y, y_hat)
        recall = recall_score(self.test_y, y_hat)
        f1 = f1_score(self.test_y, y_hat)

        self.results = {
            "accuracy": accuracy,
            "precision": precision,
            "recall": recall,
            "f1": f1,
        }

    def predict(self, text):
        pass

    def report(self):
        return self.results

    def run(self):
        self.train()
        self.test()
```

# Step 12: Define Constants & Model Configurations

```python
MAX_FEATURES = 2000
TEST_SIZE = 0.2
MAX_ITER = 1000
SEED = 100
X_COLUMNS = list(preprocessing_vocab.keys())


result_metrics = {
    "remove_stop_words": [],
    "stem": [],
    "lemmatize": [],
    "vectorizer": [],
    "classifier": [],
    "accuracy": [],
    "precision": [],
    "recall": [],
    "f1": [],
}

classifiers = {
    'MultinomialNB': MultinomialNB(),
    'BernoulliNB': BernoulliNB(),
    'LogisticRegression': LogisticRegression(max_iter=MAX_ITER),
    'SVM': LinearSVC()
}

vectorizers = {
    'tf_idf': TfidfVectorizer(max_features=MAX_FEATURES),
    'word_counter': CountVectorizer(max_features=MAX_FEATURES, binary=True)
}

```

# Step 13: Model Training & Evaluation

```python
# 80% training and 20% testing
for column in X_COLUMNS:
    train_x, test_x, train_y, test_y = train_test_split(
      df[column],
      df['target'],
      test_size=TEST_SIZE,
      random_state=SEED
    )
    for vec_name, vectorizer in vectorizers.items():
        for model_name, model in classifiers.items():
            classifier = MLModel(
                model = model,
                train_x = train_x,
                test_x = test_x,
                train_y = train_y,
                test_y = test_y,
                stemmer = PorterStemmer(),
                lemmatizer = WordNetLemmatizer(),
                tokenizer = word_tokenize,
                vectorizer = vectorizer
            )

            reporter_name = f"{getPreprocessingConfig(column)}-{vec_name}-{model_name}"

            classifier.run()
            report = classifier.report()


            result_metrics["remove_stop_words"].append(True if "remove_stop_words=True" in reporter_name else False)
            result_metrics["stem"].append(True if "stem=True" in reporter_name else False)
            result_metrics["lemmatize"].append(True if "lemmatize=True" in reporter_name else False)
            result_metrics['vectorizer'].append(vec_name)
            result_metrics['classifier'].append(model_name)
            result_metrics['accuracy'].append(report['accuracy'])
            result_metrics['precision'].append(report['precision'])
            result_metrics['recall'].append(report['recall'])
            result_metrics['f1'].append(report['f1'])
            print(f"{reporter_name} --> {report}")
```

# Models

## Naïve Bayes

### 1. Bernoulli Model
An event – a document as a whole
* a bag of words
* words are attributes of the event
* vocabulary **term ω** is a **Bernoulli attribute**
  * **1**, if **ω** is in the document
  * **0**, otherwise
<!-- * non-occurance  -->

### 2. Multinomial Model
Document – a sequence of events W1,…,W|d|  
* Considerations:
    * number of occurrences
    * length of the document
<!-- * non-occurance doesn't affect classification decision -->


## Logistic Regression
* $x_1, x_2, x_3, ... x_n$ as features
* $y = {0, 1}$ as target
$$ y^{'} = f(x) = w_1.x_1 + w_2.x_2 + w_3.x_3 + ... w_n.x_n + b $$
* sigmoid activation function is used for classification:
$$ \sigma(x) = \frac{1}{({1+e^{x}})} $$
$$ P(y = 1) = \sigma(x) $$
$$ P(y = 0) = 1 - \sigma(x) $$
And,
$$ logit(p) = \sigma^{-1}(p) = ln(\frac{1}{1-p}) $$
**Decision boundry**
  $$ decision(x) = \begin{cases} 1, & \text{if } P(y=1 \mid x) > 0.5,\\ 0, & \text{otherwise} \end{cases} $$
